```python
import sqlite3
import pandas as pd
conn = sqlite3.connect('../databases/pokedex.db')
conn.row_factory = sqlite3.Row
c = conn.cursor()
```

```python
import pandas as pd

url = "https://pokemondb.net/ability"
tables = pd.read_html(url)

df = next(t for t in tables if {"Name", "Description"}.issubset(t.columns))
ability_info = df[["Name", "Description"]]
ability_info
```

```python
c.execute("""CREATE TABLE IF NOT EXISTS Abilities (
    NAME TEXT PRIMARY KEY,
    DESCRIPTION TEXT
)""")
```

```python
ability_info.to_sql('Abilities', conn, if_exists='replace', index=False)
```

```python
c.execute("""SELECT * FROM Abilities""")
rows = c.fetchall()
for row in rows:
    print(row['NAME'], "—", row['DESCRIPTION'])
```
